```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gastos_pessoais.csv")

print("Total de registros:", len(df))

print("\nGastos por categoria:")
print(df.groupby("categoria")["valor"].sum())

df.groupby("categoria")["valor"].sum().plot(kind="bar")

plt.title("Gastos por categoria")
plt.ylabel("Valor")
plt.xlabel("Categoria")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gastos_pessoais.csv")

df["data"] = pd.to_datetime(df["data"])
df["mes"] = df["data"].dt.month

print(df.head())
```

```python
print(df.groupby("mes")["valor"].sum())

df.groupby("mes")["valor"].sum().plot(kind="bar")
plt.title("Gastos por mês")
plt.ylabel("Valor")
plt.xlabel("Mês")
plt.show()
```
